```python
from sentence_transformers import SentenceTransformer, util
encoder = SentenceTransformer("sentence-transformers/gtr-t5-base")
```

```python
encoder
```

```python
tokenizer = encoder.tokenizer
text = "Ala ma kota"
enc_text = tokenizer.encode(text)
enc_text
```

```python
vocab = tokenizer.vocab
inv_vocab = {v: k for k,v in vocab.items()}
tokens = [inv_vocab[item] for item in enc_text]
tokens
```

```python
def get_tokens(text):
    enc_text = tokenizer.encode(text)
    tokens = [inv_vocab[item] for item in enc_text]
    return tokens

sentences = ["abcd", "xdddd", "pare roznych slow oraz zdan", "and something in english :D"] 
sentences_tokens = [get_tokens(text) for text in sentences]
sentences_tokens   
```
